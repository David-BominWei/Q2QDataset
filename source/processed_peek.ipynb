```python
import pandas as pd

newdata = pd.read_csv("data.csv").set_index("idx")
newdata
```

```python
qq_dictlist = []

for idx in newdata.index:
    line = newdata.loc[idx]
    dict_line = {"q":line["q0"], "qa_candidates":[[1,line["qa"]]]}
    for ll in range(1,6):
        dict_line["qa_candidates"].append([0,line["qn"+str(ll)]])
    qq_dictlist.append(dict_line)
    
qq_dictlist
```

```python
import json

traintestbinary = int(len(qq_dictlist)*0.9)
trainset = qq_dictlist[:traintestbinary]
valset = qq_dictlist[traintestbinary:]

with open("data/train.json", "w") as F:
    json.dump(trainset, F, ensure_ascii=False, indent=4)
    
with open("data/validate.json", "w") as F:
    json.dump(valset, F, ensure_ascii=False, indent=4)
```

```python
testset = []
testdata = pd.read_csv("dev.tsv", sep="\t", header=None, names=["q0","qa","label"])

for idx in testdata.index:
    line = testdata.loc[idx]
    testline = {"q":line["q0"], "qa_candidates":[str(line["label"]),line["qa"]]}
    testset.append(testline)
    
print(testset)

with open("data/test.json", "w") as F:
    json.dump(testset, F, ensure_ascii=False, indent=4)
```
